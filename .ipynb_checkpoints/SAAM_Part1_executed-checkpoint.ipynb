# SAAM Project – Part I
**Group AD | North America + Europe | Scope 1+2**

Steps:
1. Load all the data files
2. Keep only AMER and EUR firms
3. Clean the data (missing prices, delisted firms, stale prices)
4. Build the investment set each year
5. Run min-variance optimization (rolling, annual)
6. Compute portfolio returns (MV and value-weighted)
7. Summary stats + plot
8. Fill the template Excel

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import cvxpy as cp
import re, os, shutil, warnings

warnings.filterwarnings('ignore')

# paths — adjust if needed
DATA_DIR      = os.path.join(os.path.expanduser('~'), 'Downloads', 'Data_2026 (1)')
TEMPLATE_PATH = os.path.join(os.path.expanduser('~'), 'Downloads', 'Template for Part I-SAAM (1).xlsx')
OUTPUT_PATH   = 'SAAM_Part1_Results.xlsx'

# group AD parameters
REGIONS      = ['AMER', 'EUR']
Y0           = 2013     # first rebalancing year
Y_END        = 2024     # last rebalancing year
WIN_YEARS    = 10       # estimation window in years
MIN_OBS      = 36       # min monthly return observations required (3 years)
LOW_PRICE    = 0.5      # prices below this → treat as missing
STALE_THRESH = 0.50     # if more than 50% of returns are 0 → illiquid, skip

print('data dir exists:', os.path.isdir(DATA_DIR))
```

## 1. Load data

All Datastream files have the same structure: first row is a broken error row from Datastream, then firm NAME as index, first column = ISIN, rest = time series. So I load them all the same way.

```python
def load_ds_file(filename):
    """Load a standard Datastream export. Returns (isin_series, data_df)."""
    df = pd.read_excel(os.path.join(DATA_DIR, filename), index_col=0, header=0)
    df = df[df.index.notna()]   # first row is always an error placeholder, drop it
    isin_col = df.iloc[:, 0]    # ISIN is always the first column
    data     = df.iloc[:, 1:].apply(pd.to_numeric, errors='coerce')
    return isin_col, data

# static firm info
static = pd.read_excel(os.path.join(DATA_DIR, 'Static_2025.xlsx')).set_index('ISIN')

# monthly price/market cap
isin_ri, ri   = load_ds_file('DS_RI_T_USD_M_2025.xlsx')
isin_mv, mv   = load_ds_file('DS_MV_T_USD_M_2025.xlsx')

# annual carbon + revenue
isin_c1, co2_1 = load_ds_file('DS_CO2_SCOPE_1_Y_2025.xlsx')
isin_c2, co2_2 = load_ds_file('DS_CO2_SCOPE_2_Y_2025.xlsx')
isin_rv, rev   = load_ds_file('DS_REV_Y_2025.xlsx')

# risk-free rate (monthly, in %)
rf_raw = pd.read_excel(os.path.join(DATA_DIR, 'Risk_Free_Rate_2025.xlsx'),
                        header=None, names=['yyyymm', 'rf']).dropna()
rf_raw['date'] = pd.to_datetime(rf_raw['yyyymm'].astype(int).astype(str),
                                 format='%Y%m') + pd.offsets.MonthEnd(0)
rf = rf_raw.set_index('date')['rf'].astype(float) / 100   # convert to decimal

# convert monthly column headers to proper datetime
ri.columns = pd.to_datetime(ri.columns)
mv.columns = pd.to_datetime(mv.columns)

print('RI shape:', ri.shape)
print('MV shape:', mv.shape)
print('CO2 scope 1 shape:', co2_1.shape)
```

## 2. Filter to AMER + EUR

```python
# get all ISINs that belong to our regions
target_isins = set(static[static['Region'].isin(REGIONS)].index)
print(f'Total firms in AMER+EUR: {len(target_isins)}')

# filter each panel to only keep those firms
mask_ri = isin_ri.isin(target_isins)
ri = ri[mask_ri];   isin_ri = isin_ri[mask_ri]

mask_mv = isin_mv.isin(target_isins)
mv = mv[mask_mv];   isin_mv = isin_mv[mask_mv]

mask_c1 = isin_c1.isin(target_isins)
co2_1 = co2_1[mask_c1];   isin_c1 = isin_c1[mask_c1]

mask_c2 = isin_c2.isin(target_isins)
co2_2 = co2_2[mask_c2];   isin_c2 = isin_c2[mask_c2]

mask_rv = isin_rv.isin(target_isins)
rev = rev[mask_rv];   isin_rv = isin_rv[mask_rv]

print(f'After filter — RI: {ri.shape}, MV: {mv.shape}')
```

## 3. Data cleaning

Three things to fix before computing returns:
- **Delisted firms**: Datastream puts `DEAD - DELIST.DD/MM/YY` in the name. We set the last price to 0 (= -100% return) and everything after to NaN.
- **Very low prices**: RI below 0.5 causes garbage returns (division by near-zero). Treat as NaN.
- **Carbon data gaps**: forward-fill missing years, but only from the first year the firm reported.

```python
# regex to pull the delisting date out of the firm name
DELIST_PATTERN = re.compile(r'DEAD\s*-\s*(?:DELIST|DEAD)\.?(\d{2}/\d{2}/\d{2,4})', re.IGNORECASE)

def get_delist_date(name):
    match = DELIST_PATTERN.search(str(name))
    if not match:
        return None
    date_str = match.group(1)
    fmt = '%d/%m/%y' if len(date_str.split('/')[-1]) == 2 else '%d/%m/%Y'
    try:
        return pd.to_datetime(date_str, format=fmt) + pd.offsets.MonthEnd(0)
    except:
        return None

# start from a clean copy
ri_clean = ri.copy().astype(float)

# low prices → NaN
ri_clean[ri_clean < LOW_PRICE] = np.nan

# handle delisted firms
n_delisted = 0
for name in ri.index:
    ddate = get_delist_date(name)
    if ddate is None:
        continue
    n_delisted += 1
    # set everything after delist to NaN
    after_delist = ri_clean.columns[ri_clean.columns > ddate]
    ri_clean.loc[name, after_delist] = np.nan
    # set the last available price to 0 so return = -100%
    up_to_delist = ri_clean.columns[ri_clean.columns <= ddate]
    if len(up_to_delist) > 0:
        ri_clean.loc[name, up_to_delist[-1]] = 0.0

print(f'Delisted firms found: {n_delisted}')

# compute simple monthly returns: R_t = P_t/P_{t-1} - 1
# important: fill_method=None so we don't forward-fill prices before computing returns
ret = ri_clean.pct_change(axis=1, fill_method=None).iloc[:, 1:]
ret = ret.clip(lower=-1.0)   # cap at -100%

print('Returns matrix shape:', ret.shape)
print('Date range:', ret.columns[0].date(), '→', ret.columns[-1].date())
```

```python
# forward-fill CO2 and revenue data
# rule: only fill from the first year the firm has data (no backfill)

def ffill_annual(df):
    out = df.copy().astype(float)
    for name in out.index:
        row = out.loc[name]
        first_valid = row.first_valid_index()
        if first_valid is not None:
            out.loc[name, first_valid:] = row.loc[first_valid:].ffill()
    return out

co2_1 = ffill_annual(co2_1)
co2_2 = ffill_annual(co2_2)
rev    = ffill_annual(rev)

# total CO2 = scope 1 + scope 2, aligned to the return matrix firms
co2_total = co2_1.add(co2_2.reindex(co2_1.index), fill_value=0).reindex(ri.index)

print('CO2 total shape:', co2_total.shape)
# quick check: how many firms have CO2 data in 2020?
print('Firms with CO2 data in 2020:', co2_total[2020].notna().sum())
```

## 4. Investment set + min-variance optimization (rolling)

Each year from 2013 to 2024:
1. Figure out which firms we can invest in (4 criteria)
2. Estimate the covariance matrix from the past 10 years of monthly returns
3. Solve the min-variance QP
4. Store the weights

**Note on dates**: Datastream stores the last *trading* day of the month, not always the 31st. For example December 2016 = Dec 30, December 2017 = Dec 29. So I can't just use `Timestamp(Y, 12, 31)` — I need to find the actual December column in the data.

```python
def get_dec(Y, cols):
    """Get the actual last December date for year Y from the column index."""
    dec_cols = cols[(cols.year == Y) & (cols.month == 12)]
    return dec_cols[-1] if len(dec_cols) > 0 else None


def get_eligible_firms(Y):
    """Returns list of firm names eligible for investment in year Y+1."""
    dec_Y = get_dec(Y, ret.columns)
    if dec_Y is None:
        return []

    # 10-year window: Jan of (Y-9) to Dec of Y
    dec_Y_minus_10 = get_dec(Y - 10, ret.columns)
    window = ret.columns[(ret.columns > (dec_Y_minus_10 or ret.columns[0])) &
                          (ret.columns <= dec_Y)]
    ret_window = ret[window]

    dec_Y_price = get_dec(Y, ri_clean.columns)
    if dec_Y_price is None:
        return []

    eligible = []
    for name in ri.index:
        row = ret_window.loc[name]

        # need at least 3 years of monthly data
        if row.notna().sum() < MIN_OBS:
            continue

        # price must be valid (not NaN and not too low) at end of Y
        price = ri_clean.loc[name, dec_Y_price]
        if pd.isna(price) or price < LOW_PRICE:
            continue

        # skip stale stocks (more than 50% of returns are exactly 0)
        valid_rets = row.dropna()
        if len(valid_rets) > 0:
            if (valid_rets == 0).sum() / len(valid_rets) > STALE_THRESH:
                continue

        # must have CO2 data available (for Part II consistency)
        available_years = [c for c in co2_total.columns
                           if isinstance(c, (int, np.integer)) and c <= Y]
        if not available_years:
            continue
        last_co2_year = max(available_years)
        if name not in co2_total.index or pd.isna(co2_total.loc[name, last_co2_year]):
            continue

        eligible.append(name)

    return eligible


def estimate_cov(names, Y):
    """Estimate covariance matrix from 10 years of monthly returns ending Dec Y."""
    dec_Y        = get_dec(Y, ret.columns)
    dec_Y_minus_10 = get_dec(Y - 10, ret.columns)
    window       = ret.columns[(ret.columns > (dec_Y_minus_10 or ret.columns[0])) &
                                (ret.columns <= dec_Y)]
    sub = ret.loc[names, window].T   # T x N
    sub = sub.fillna(sub.mean())     # fill gaps with the cross-sectional mean that month
    T   = len(sub)
    mu  = sub.mean().values
    cov = ((sub.values - mu).T @ (sub.values - mu)) / T   # ML estimator
    return cov


def solve_min_var(cov):
    """Solve long-only min-variance: min w'Cov*w s.t. sum(w)=1, w>=0."""
    # I originally tried scipy.optimize.minimize with SLSQP but it was way too slow
    # (~600s for 800+ stocks). CVXPY with CLARABEL solves the same problem in <1s.
    N = cov.shape[0]
    w = cp.Variable(N)
    problem = cp.Problem(
        cp.Minimize(cp.quad_form(w, cov)),
        [cp.sum(w) == 1, w >= 0]
    )
    problem.solve(solver=cp.CLARABEL)

    if problem.status not in ('optimal', 'optimal_inaccurate') or w.value is None:
        # fallback to equal weights if something goes wrong
        return np.ones(N) / N

    weights = np.maximum(w.value, 0)   # clip tiny negatives from numerical noise
    return weights / weights.sum()


# --- main loop ---
# for each year: get eligible firms, estimate cov, solve, store weights
all_weights = {}   # {year: pd.Series(weights, index=firm_names)}

for Y in range(Y0, Y_END + 1):
    firms = get_eligible_firms(Y)

    if len(firms) == 0:
        print(f'{Y}: no eligible firms')
        continue

    cov     = estimate_cov(firms, Y)
    weights = solve_min_var(cov)
    all_weights[Y] = pd.Series(weights, index=firms)

    n_active = (weights > 1e-5).sum()
    vol_est  = np.sqrt(weights @ cov @ weights * 12) * 100
    print(f'{Y}: {len(firms)} firms eligible, {n_active} get nonzero weight, est. vol = {vol_est:.2f}%')

print('\ndone.')
```

## 5. Compute monthly portfolio returns (2014–2025)

For the MV portfolio, weights are set at the end of year Y and then **drift** during year Y+1 as prices move (we don't rebalance monthly). The drift formula is:

$$w_{i,t+1} = w_{i,t} \times \frac{1 + R_{i,t+1}}{1 + R_{p,t+1}}$$

For VW, weights are just market cap of each firm divided by total market cap, recomputed each month.

```python
# out-of-sample period: Jan 2014 to Dec 2025
oos = ret.columns[(ret.columns >= '2014-01-01') & (ret.columns <= '2025-12-31')]
print(f'OOS: {oos[0].date()} → {oos[-1].date()}, {len(oos)} months')

# --- min-variance returns ---
mv_returns = []

for Y in range(Y0, Y_END + 1):
    if Y not in all_weights:
        continue

    w = all_weights[Y].copy()     # weights decided at end of Y
    firms = w.index.tolist()
    months_next_year = [t for t in oos if t.year == Y + 1]

    for t in months_next_year:
        # get firm returns this month (NaN → 0, firm just not trading)
        r_firms = ret.loc[firms, t].fillna(0)
        r_port  = float((w * r_firms).sum())
        mv_returns.append({'date': t, 'r_mv': r_port})

        # drift-adjust weights for next month
        if abs(1 + r_port) > 1e-10:
            w = w * (1 + r_firms) / (1 + r_port)

mv_ret = pd.DataFrame(mv_returns).set_index('date')['r_mv'].reindex(oos)

# --- value-weighted returns ---
vw_returns = []

for t in oos:
    # use previous month's market cap to avoid look-ahead
    prev_dates = mv.columns[mv.columns < t]
    if len(prev_dates) == 0:
        vw_returns.append({'date': t, 'r_vw': np.nan})
        continue

    caps = mv[prev_dates[-1]].astype(float).replace(0, np.nan).dropna()
    r_t  = ret[t].dropna()
    both = caps.index.intersection(r_t.index)

    if len(both) == 0:
        vw_returns.append({'date': t, 'r_vw': np.nan})
        continue

    w_vw = caps[both] / caps[both].sum()
    vw_returns.append({'date': t, 'r_vw': float((w_vw * r_t[both]).sum())})

vw_ret = pd.DataFrame(vw_returns).set_index('date')['r_vw'].reindex(oos)

print('MV NaN months:', mv_ret.isna().sum())
print('VW NaN months:', vw_ret.isna().sum())
```

## 6. Summary statistics

```python
rf_oos = rf.reindex(oos).fillna(rf.mean())

def get_stats(returns, rf_monthly=rf_oos):
    r  = returns.dropna()
    rf_m = rf_monthly.reindex(r.index).fillna(rf_monthly.mean())
    return {
        'Annualized average return':    r.mean() * 12,
        'Annualized volatility':        r.std() * np.sqrt(12),
        'Annualized cumulative return': (1 + r).prod() ** (12 / len(r)) - 1,
        'Sharpe ratio':                 (r - rf_m).mean() / r.std() * np.sqrt(12),
        'Minimum monthly return':       r.min(),
        'Maximum monthly return':       r.max(),
    }

stats_vw = get_stats(vw_ret)
stats_mv = get_stats(mv_ret)

results = pd.DataFrame({'Value-Weighted': stats_vw, 'Min-Variance': stats_mv})
display(results.style.format('{:.4f}').set_caption('Summary Statistics 2014–2025'))
```

## 7. Plot cumulative returns

```python
cum_vw = (1 + vw_ret.fillna(0)).cumprod()
cum_mv = (1 + mv_ret.fillna(0)).cumprod()

fig, ax = plt.subplots(figsize=(12, 5))
ax.plot(cum_vw.index, cum_vw.values, label='Value-Weighted (benchmark)', color='steelblue', lw=1.8)
ax.plot(cum_mv.index, cum_mv.values, label='Min-Variance',               color='firebrick',  lw=1.8)
ax.set_title('Cumulative Returns — AMER + EUR, 2014–2025 (Group AD)', fontsize=13)
ax.set_ylabel('Cumulative return (base = 1)')
ax.xaxis.set_major_locator(mdates.YearLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
ax.legend()
ax.grid(alpha=0.3)
plt.tight_layout()
plt.savefig('cumulative_returns_part1.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 8. Fill the template Excel

```python
from openpyxl import load_workbook

shutil.copy(TEMPLATE_PATH, OUTPUT_PATH)
wb = load_workbook(OUTPUT_PATH)
ws = wb.active

# summary stats: col B = VW, col C = MV, rows 3 to 8
stat_keys = [
    'Annualized average return',
    'Annualized volatility',
    'Annualized cumulative return',
    'Sharpe ratio',
    'Minimum monthly return',
    'Maximum monthly return',
]
for i, key in enumerate(stat_keys):
    r = 3 + i
    ws.cell(row=r, column=2).value = round(stats_vw[key], 8)
    ws.cell(row=r, column=3).value = round(stats_mv[key], 8)

# monthly returns: col F = VW, col G = MV, starting row 3
for i, t in enumerate(oos):
    r  = 3 + i
    vv = vw_ret.get(t, np.nan)
    mv_val = mv_ret.get(t, np.nan)
    ws.cell(row=r, column=6).value = None if np.isnan(float(vv))     else round(float(vv), 8)
    ws.cell(row=r, column=7).value = None if np.isnan(float(mv_val)) else round(float(mv_val), 8)

wb.save(OUTPUT_PATH)
print(f'saved → {OUTPUT_PATH}')
```
