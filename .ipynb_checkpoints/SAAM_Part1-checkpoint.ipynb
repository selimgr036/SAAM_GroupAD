# SAAM Part I — Group AD (AMER + EUR, Scope 1 + Scope 2)

```python
import pandas as pd
import numpy as np
import cvxpy as cp
import matplotlib.pyplot as plt
import seaborn as sns
import re, os, shutil, warnings

warnings.filterwarnings('ignore')

# Load raw data from Excel
fichier_firm       = "Static_2025.xlsx"
fichier_CO2_SCOPE1 = "DS_CO2_SCOPE_1_Y_2025.xlsx"
fichier_CO2_SCOPE2 = "DS_CO2_SCOPE_2_Y_2025.xlsx"
fichier_REV_USD_Y  = "DS_REV_Y_2025.xlsx"
fichier_MC_USD_M   = "DS_MV_T_USD_M_2025.xlsx"
fichier_TRI_USD_M  = "DS_RI_T_USD_M_2025.xlsx"
fichier_RF_M       = "Risk_Free_Rate_2025.xlsx"
fichier_template   = "Template for Part I-SAAM (1).xlsx"
output_path        = "SAAM_Part1_Results.xlsx"

# Datastream files have a broken first row (Thomson Reuters metadata) and NAME as
# row index with ISIN as first data column — we clean this on load
def read_ds(path):
    df = pd.read_excel(path, index_col=0, header=0)
    df = df[df.index.notna()]                              # drop metadata row
    df.index = df.iloc[:, 0].values                        # ISIN → new index
    df.index.name = "ISIN"
    df = df.iloc[:, 1:].apply(pd.to_numeric, errors='coerce')
    return df

df_firm      = pd.read_excel(fichier_firm)
df_scope1    = read_ds(fichier_CO2_SCOPE1)
df_scope2    = read_ds(fichier_CO2_SCOPE2)
df_rev_usd_y = read_ds(fichier_REV_USD_Y)
df_mc_usd_m  = read_ds(fichier_MC_USD_M)
df_tri_usd_m = read_ds(fichier_TRI_USD_M)
df_rf_m      = pd.read_excel(fichier_RF_M, header=None, names=['yyyymm', 'rf']).dropna()

# Keep only firms from AMER and EUR regions (Group AD)
isin_ad = df_firm[df_firm["Region"].isin(["AMER", "EUR"])]["ISIN"].unique()

df_scope1_ad    = df_scope1[df_scope1.index.isin(isin_ad)].copy().reset_index()
df_scope2_ad    = df_scope2[df_scope2.index.isin(isin_ad)].copy().reset_index()
df_rev_usd_y_ad = df_rev_usd_y[df_rev_usd_y.index.isin(isin_ad)].copy().reset_index()
df_mc_usd_m_ad  = df_mc_usd_m[df_mc_usd_m.index.isin(isin_ad)].copy().reset_index()
df_tri_usd_m_ad = df_tri_usd_m[df_tri_usd_m.index.isin(isin_ad)].copy().reset_index()
df_firm_ad      = df_firm[df_firm["ISIN"].isin(isin_ad)].copy().reset_index(drop=True)

# Clean each DataFrame in the list — set ISIN as index and strip column names
df_list = [df_scope1_ad, df_scope2_ad, df_rev_usd_y_ad, df_mc_usd_m_ad, df_tri_usd_m_ad]

for i, df in enumerate(df_list):
    if "ISIN" in df.columns:
        df = df.set_index("ISIN")
    df.columns = [str(col).strip() for col in df.columns]
    df_list[i] = df

(df_scope1_ad, df_scope2_ad, df_rev_usd_y_ad, df_mc_usd_m_ad, df_tri_usd_m_ad) = df_list

# Convert column types
df_mc_usd_m_ad.columns  = pd.to_datetime(df_mc_usd_m_ad.columns, errors='coerce')
df_tri_usd_m_ad.columns = pd.to_datetime(df_tri_usd_m_ad.columns, errors='coerce')

df_rev_usd_y_ad.columns = df_rev_usd_y_ad.columns.astype(int)
df_scope1_ad.columns    = df_scope1_ad.columns.astype(int)
df_scope2_ad.columns    = df_scope2_ad.columns.astype(int)

# Remove NaT columns that may have slipped through
df_mc_usd_m_ad  = df_mc_usd_m_ad[[c for c in df_mc_usd_m_ad.columns if not pd.isna(c)]]
df_tri_usd_m_ad = df_tri_usd_m_ad[[c for c in df_tri_usd_m_ad.columns if not pd.isna(c)]]

# Check alignment
print(" ANNUAL: Scope 1 vs Scope 2")
print("Same columns:", all(df_scope1_ad.columns == df_scope2_ad.columns))

print("\n MONTHLY: TRI vs Market Cap")
print("Same columns:", all(df_tri_usd_m_ad.columns == df_mc_usd_m_ad.columns))

# Fill missing values and remove firms with 2+ consecutive missing years
def fill_and_count(df, name):
    df = df.ffill(axis=1).infer_objects(copy=False)

    def has_two_consecutive_nans(row):
        first_valid = row.first_valid_index()
        if first_valid is None:
            return True
        row_after_entry = row.loc[first_valid:]
        is_nan = row_after_entry.isna()
        count = 0
        for val in is_nan:
            if val:
                count += 1
                if count >= 2:
                    return True
            else:
                count = 0
        return False

    mask_valid = ~df.apply(has_two_consecutive_nans, axis=1)
    df = df.loc[mask_valid]
    print(f"{name} → Firms removed: {len(mask_valid) - mask_valid.sum()} | Remaining: {df.shape[0]}")
    return df

df_scope1_ad    = fill_and_count(df_scope1_ad,    "df_scope1_ad   ")
df_scope2_ad    = fill_and_count(df_scope2_ad,    "df_scope2_ad   ")
df_rev_usd_y_ad = fill_and_count(df_rev_usd_y_ad, "df_rev_usd_y_ad")

common_firms = (df_scope1_ad.index
                .intersection(df_scope2_ad.index)
                .intersection(df_rev_usd_y_ad.index))

print(f"\nNumber of common firms: {len(common_firms)}")

# Filter all DataFrames to common firms
df_scope1_ad    = df_scope1_ad.loc[common_firms]
df_scope2_ad    = df_scope2_ad.loc[common_firms]
df_rev_usd_y_ad = df_rev_usd_y_ad.loc[common_firms]
df_mc_usd_m_ad  = df_mc_usd_m_ad.loc[df_mc_usd_m_ad.index.isin(common_firms)]
df_tri_usd_m_ad = df_tri_usd_m_ad.loc[df_tri_usd_m_ad.index.isin(common_firms)]
df_firm_ad      = df_firm_ad[df_firm_ad["ISIN"].isin(common_firms)]

# Align risk-free rate index with monthly return dates
df_rf_m['date'] = pd.to_datetime(df_rf_m['yyyymm'].astype(int).astype(str),
                                  format='%Y%m') + pd.offsets.MonthEnd(0)
df_rf_m = df_rf_m.set_index('date')['rf'].astype(float) / 100
df_rf_m = df_rf_m[df_rf_m.index >= pd.Timestamp("2014-01-01")]

# Convert revenues from kUSD to MUSD
df_rev_usd_y_ad = df_rev_usd_y_ad.apply(pd.to_numeric, errors='coerce') / 1000

# Total emissions = Scope 1 + Scope 2
df_emissions_ad = df_scope1_ad + df_scope2_ad.reindex(df_scope1_ad.index)

# Carbon intensity (CI) = emissions / revenue
df_ci_ad = df_emissions_ad / df_rev_usd_y_ad

# Calculate monthly returns from Total Return Index
df_returns_usd_m_ad = df_tri_usd_m_ad.T.pct_change().T
df_returns_usd_m_ad.replace([np.inf, -np.inf], np.nan, inplace=True)

print("\n MONTHLY: TRI vs Returns")
print("Columns identical:", all(df_tri_usd_m_ad.columns == df_returns_usd_m_ad.columns))

# Country distribution
country_pct = df_firm_ad["Country"].value_counts(normalize=True) * 100
country_pct_df = country_pct.round(2).reset_index()
country_pct_df.columns = ["Country", "Percentage"]
print("\nDistribution of companies by country:")
print(country_pct_df.head(15))

# Filter delisted firms from firm list
mask = df_firm_ad['NAME'].str.contains(r'DELIST\.\d{2}/\d{2}/\d{2}', regex=True, na=False)
delisted_firms = df_firm_ad.loc[mask, ['ISIN', 'NAME']].copy()
delisted_firms['Delist_Date'] = delisted_firms['NAME'].str.extract(r'DELIST\.(\d{2}/\d{2}/\d{2})')
delisted_firms['Delist_Date'] = pd.to_datetime(delisted_firms['Delist_Date'],
                                               format='%d/%m/%y', errors='coerce')
delisted_firms['Delisted_Year'] = delisted_firms['Delist_Date'].dt.year
delisted_firms = delisted_firms.set_index('ISIN')[['Delisted_Year']].dropna()

print(f"\nDelisted firms detected: {len(delisted_firms)}")

# Define rolling window (10 years) and rebalancing frequency (annual)
window_months  = 120
rebalance_step = 12

# Note: Datastream stores the last TRADING day of the month (not always Dec 31).
# e.g. Dec 2016 = Dec 30, Dec 2017 = Dec 29. Using Timestamp(Y,12,31) caused
# empty investment sets — we search the column index for the actual December date.
def get_dec(cols, year):
    dec = [c for c in cols if c.year == year and c.month == 12]
    return dec[-1] if dec else None

start_date  = get_dec(df_returns_usd_m_ad.columns, 2013)
end_date    = get_dec(df_returns_usd_m_ad.columns, 2024)
start_index = list(df_returns_usd_m_ad.columns).index(start_date)
end_index   = list(df_returns_usd_m_ad.columns).index(end_date)

selected_firms_by_year = {}

for t in range(start_index, end_index + 1, rebalance_step):

    date_rebalance = df_returns_usd_m_ad.columns[t]
    year = date_rebalance.year

    # Data recovery
    R_past = df_returns_usd_m_ad.iloc[:, t - window_months:t]
    R_next = df_returns_usd_m_ad.iloc[:, t:t + rebalance_step]
    E_y    = df_emissions_ad[year] if year in df_emissions_ad.columns else None
    cap_m  = df_mc_usd_m_ad.iloc[:, t:t + rebalance_step]

    # Apply filters
    mask_past      = R_past.notna().sum(axis=1) >= 60
    mask_next      = R_next.notna().all(axis=1)
    mask_emissions = E_y.notna() if E_y is not None else pd.Series(False, index=R_past.index)
    mask_cap_m     = cap_m.notna().all(axis=1)

    mask_total  = mask_past & mask_emissions & mask_next & mask_cap_m
    valid_firms = mask_total[mask_total].index

    # Exclude firms delisted this year or before
    if not delisted_firms.empty:
        isins_excluded = delisted_firms[delisted_firms['Delisted_Year'] <= year].index
        valid_firms = valid_firms.difference(isins_excluded)

    selected_firms_by_year[date_rebalance] = valid_firms.tolist()
    print(f"{date_rebalance.date()} → {len(valid_firms)} firms eligible")

    # Minimum Variance Portfolio (MVP)

# We minimize portfolio variance: min w'Σw  s.t.  Σw=1, w≥0
# scipy SLSQP timed out (>600s) for 800+ stocks — replaced with CVXPY CLARABEL

returns_mvp_m       = []
weights_mvp_by_year = {}

for t in range(start_index, end_index + 1, rebalance_step):

    date_rebalance = df_returns_usd_m_ad.columns[t]
    year = date_rebalance.year

    # Data recovery
    R_past = df_returns_usd_m_ad.iloc[:, t - window_months:t]
    R_next = df_returns_usd_m_ad.iloc[:, t:t + rebalance_step]

    # Recovery of registered valid firms
    valid_firms = selected_firms_by_year[date_rebalance]

    # Application
    R_past = R_past.loc[valid_firms]
    R_next = R_next.loc[valid_firms]

    print(f"{date_rebalance.date()} → {R_past.shape[0]} firms used")

    # Compute covariance matrix from past data (fill NaN with cross-sectional mean)
    R_past_filled = R_past.T.fillna(R_past.T.mean())
    Sigma_hat     = R_past_filled.cov().to_numpy()

    # Quadratic program
    n  = len(valid_firms)
    w  = cp.Variable(n)

    constraint_budget = [cp.sum(w) == 1, w >= 0]

    def objective(Sigma): return cp.quad_form(w, Sigma)

    prob = cp.Problem(cp.Minimize(objective(Sigma_hat)), constraint_budget)
    prob.solve(solver=cp.CLARABEL)

    if prob.status not in ('optimal', 'optimal_inaccurate') or w.value is None:
        print(f"  Optimization failed for {date_rebalance.date()}")
        continue

    w_opt = np.maximum(w.value, 0)
    w_opt = w_opt / w_opt.sum()

    # Save weights
    weights_mvp_by_year[date_rebalance] = pd.Series(w_opt, index=R_past.index)

    # Compute portfolio returns (Buy-and-Hold with annual drift adjustment)
    w_t = w_opt.copy()
    returns_mvp_year = []

    for i, date in enumerate(R_next.columns):
        r_i = R_next[date].values
        r_p = np.dot(w_t, r_i)

        returns_mvp_year.append(r_p)

        # Weight update (Buy-and-Hold)
        w_t = w_t * (1 + r_i) / (1 + r_p)

    returns_mvp_m.extend(returns_mvp_year)

# Reconstruct the datetime index for the portfolio returns
dates_ret           = df_returns_usd_m_ad.columns[start_index:start_index + len(returns_mvp_m)]
returns_mvp_m_series = pd.Series(returns_mvp_m, index=dates_ret)
weights_mvp_by_year  = pd.DataFrame(weights_mvp_by_year)

# Compute performance statistics
rf_aligned  = df_rf_m.reindex(returns_mvp_m_series.index).fillna(df_rf_m.mean())
mu_mvp_y    = returns_mvp_m_series.mean() * 12
vol_mvp_y   = returns_mvp_m_series.std()  * np.sqrt(12)
sharpe_mvp  = (mu_mvp_y - rf_aligned.mean() * 12) / vol_mvp_y
min_ret_mvp = returns_mvp_m_series.min()
max_ret_mvp = returns_mvp_m_series.max()
cum_ret_mvp = (1 + returns_mvp_m_series).prod() ** (12 / len(returns_mvp_m_series)) - 1

performance_summary_mvp = pd.DataFrame({
    "Minimum variance portfolio": [
        mu_mvp_y, vol_mvp_y, cum_ret_mvp, sharpe_mvp, min_ret_mvp, max_ret_mvp
    ]
}, index=[
    "Annualized average return",
    "Annualized volatility",
    "Annualized cumulative return",
    "Sharpe ratio",
    "Minimum monthly return",
    "Maximum monthly return"
])

print("\nPerformance Summary:")
print(performance_summary_mvp)

# Cumulative return of MVP
cum_mvp = (1 + returns_mvp_m_series).cumprod()

plt.figure(figsize=(12, 5))
plt.plot(cum_mvp, label="MVP", color="blue", linewidth=2)
plt.title("Cumulative Return – Minimum Variance Portfolio", fontsize=14)
plt.xlabel("Year")
plt.ylabel("Cumulative Return")
plt.grid(True, linestyle="--", linewidth=0.6)
plt.legend()
plt.tight_layout()
plt.show()

    # Benchmark Portfolio (Value-Weighted)

returns_bmk_m       = []
weights_bmk_by_year = {}

for t in range(start_index, end_index + 1, rebalance_step):

    date_rebalance = df_returns_usd_m_ad.columns[t]
    year = date_rebalance.year

    # Data recovery
    R_next = df_returns_usd_m_ad.iloc[:, t:t + rebalance_step]
    cap_t  = df_mc_usd_m_ad[date_rebalance]

    # Recovery of registered valid firms
    valid_firms = selected_firms_by_year[date_rebalance]

    # Application
    R_next = R_next.loc[valid_firms]
    cap_t  = cap_t.loc[valid_firms].astype(float).replace(0, np.nan).dropna()
    valid_firms_cap = cap_t.index.intersection(R_next.index)
    cap_t  = cap_t.loc[valid_firms_cap]
    R_next = R_next.loc[valid_firms_cap]

    print(f"{date_rebalance.date()} → {R_next.shape[0]} firms used")

    w_cap = cap_t / cap_t.sum()
    weights_bmk_by_year[date_rebalance] = pd.Series(w_cap.values, index=R_next.index)

    w_t = w_cap.values.copy()
    returns_bmk_year = []

    for i, date in enumerate(R_next.columns):
        r_i = R_next[date].values
        r_p = np.dot(w_t, r_i)

        returns_bmk_year.append(r_p)

        # Weight update (Buy-and-Hold)
        w_t = w_t * (1 + r_i) / (1 + r_p)

    returns_bmk_m.extend(returns_bmk_year)

# Reconstruct the datetime index for the portfolio returns
returns_bmk_m_series = pd.Series(returns_bmk_m, index=dates_ret)
weights_bmk_by_year  = pd.DataFrame(weights_bmk_by_year)

# Compute performance statistics
rf_aligned  = df_rf_m.reindex(returns_bmk_m_series.index).fillna(df_rf_m.mean())
mu_bmk_y    = returns_bmk_m_series.mean() * 12
vol_bmk_y   = returns_bmk_m_series.std()  * np.sqrt(12)
sharpe_bmk  = (mu_bmk_y - rf_aligned.mean() * 12) / vol_bmk_y
min_ret_bmk = returns_bmk_m_series.min()
max_ret_bmk = returns_bmk_m_series.max()
cum_ret_bmk = (1 + returns_bmk_m_series).prod() ** (12 / len(returns_bmk_m_series)) - 1

performance_summary_bmk = pd.DataFrame({
    "Value-weighted portfolio": [
        mu_bmk_y, vol_bmk_y, cum_ret_bmk, sharpe_bmk, min_ret_bmk, max_ret_bmk
    ]
}, index=[
    "Annualized average return",
    "Annualized volatility",
    "Annualized cumulative return",
    "Sharpe ratio",
    "Minimum monthly return",
    "Maximum monthly return"
])

print("\nPerformance Summary:")
print(performance_summary_bmk)

# Combined performance table
performance_summary_all = pd.concat([performance_summary_bmk, performance_summary_mvp], axis=1)
print("\nPerformance Summary — VW vs MVP:")
print(performance_summary_all)

# Cumulative return: BMK
cum_bmk = (1 + returns_bmk_m_series).cumprod()

plt.figure(figsize=(12, 5))
plt.plot(cum_bmk, label="Benchmark (VW)", color="green", linestyle="--", linewidth=2)
plt.title("Cumulative Return – Benchmark (Value-Weighted)", fontsize=14)
plt.xlabel("Year")
plt.ylabel("Cumulative Return")
plt.grid(True, linestyle="--", linewidth=0.6)
plt.legend()
plt.tight_layout()
plt.show()

# Cumulative return: MVP vs Benchmark
plt.figure(figsize=(12, 5))
plt.plot(cum_mvp, label="MVP", color="blue", linewidth=2)
plt.plot(cum_bmk, label="Benchmark (VW)", color="green", linestyle="--", linewidth=2)
plt.title("Cumulative Return – MVP vs Benchmark", fontsize=14)
plt.xlabel("Year")
plt.ylabel("Cumulative Return")
plt.legend()
plt.grid(True, linestyle="--", linewidth=0.6)
plt.tight_layout()
plt.savefig("cumulative_returns_part1.png", dpi=150, bbox_inches="tight")
plt.show()


    # Fill Excel template

from openpyxl import load_workbook

shutil.copy(fichier_template, output_path)
wb = load_workbook(output_path)
ws = wb.active

# Summary stats: rows 3-8, col B = VW, col C = MVP
stat_rows = [
    (mu_bmk_y,    mu_mvp_y),
    (vol_bmk_y,   vol_mvp_y),
    (cum_ret_bmk, cum_ret_mvp),
    (sharpe_bmk,  sharpe_mvp),
    (min_ret_bmk, min_ret_mvp),
    (max_ret_bmk, max_ret_mvp),
]
for i, (vw_val, mv_val) in enumerate(stat_rows):
    ws.cell(row=3 + i, column=2).value = round(vw_val, 8)
    ws.cell(row=3 + i, column=3).value = round(mv_val, 8)

# Monthly returns: col F = VW, col G = MVP, starting row 3
for i, date in enumerate(dates_ret):
    vw_val = returns_bmk_m_series.get(date, np.nan)
    mv_val = returns_mvp_m_series.get(date, np.nan)
    ws.cell(row=3 + i, column=6).value = None if np.isnan(float(vw_val)) else round(float(vw_val), 8)
    ws.cell(row=3 + i, column=7).value = None if np.isnan(float(mv_val)) else round(float(mv_val), 8)

wb.save(output_path)
print(f"Template filled and saved → {output_path}")
```
